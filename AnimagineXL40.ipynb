```python
#@title Preparation
!pip install diffusers transformers accelerate safetensors --upgrade
import datetime
import torch
from diffusers import StableDiffusionXLPipeline

pipe = StableDiffusionXLPipeline.from_pretrained(
    "cagliostrolab/animagine-xl-4.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    custom_pipeline="lpw_stable_diffusion_xl",
    add_watermarker=False
)
pipe.to('cuda')

image_no = 0
```

```python
#@title Generation
quality_prompt = "masterpiece, high score, great score, absurdres, explicit" #@param {type:"string"}
artist_prompt = "masami chie" #@param {type:"string"}
main_prompt = "1girl, arima kana, oshi no ko, hoshimachi suisei, hoshimachi suisei \\(1st costume\\), cosplay, looking at viewer, smile, outdoors, night, v, " #@param {type:"string"}
negative_prompt = "lowres, bad anatomy, bad hands, text, error, missing finger, extra digits, fewer digits, cropped, worst quality, low quality, low score, bad score, average score, signature, watermark, username, blurry" #@param {type:"string"}

width = 832 #@param
height = 1216 #@param
guidance_scale = 6 #@param
num_inference_steps = 25 #@param

image_name = f"{datetime.date.today().strftime('%Y%m%d')}-{image_no:04}"

prompt = f"{quality_prompt}, {artist_prompt}, {main_prompt}"

image = pipe(
    prompt,
    negative_prompt=negative_prompt,
    width=width,
    height=height,
    guidance_scale=guidance_scale,
    num_inference_steps=num_inference_steps
).images[0]

image.save(f"./{image_name}.png")
image_no += 1
```
